# Simple Linear Regression

## Importing the Libaries

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

## Importing the Dataset

```python
dataset = pd.read_csv('Salary_Data.csv')
X = dataset.iloc[:, :-1].values
y = dataset.iloc[:, -1].values
```

```python
print(dataset)
```

```python
print(X)
```

```python
print(y)
```

## Splitting the dataset into the Training set and Test set

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 1/3, random_state = 0)
```

## Training the Sample Linear Regression Model on the Training Set

```python
from sklearn.linear_model import LinearRegression
regressor = LinearRegression()
regressor.fit(X_train, y_train)
```

## Predicting the Test Set Results

```python
y_pred = regressor.predict(X_test)
```

```python
print(y_pred)
```

## Visualizaing the Training Test Results

```python
plt.scatter(X_train, y_train, color = 'red')
plt.plot(X_train, regressor.predict(X_train), color = 'blue')
plt.title('Salary vs. Experience (Training Set)')
plt.xlabel('Years of Experience')
plt.ylabel('Salary')
plt.show()
```

The scatter plot shows the actual salaries (red dots) plotted against years of experience, along with the regression line (blue line) representing the model's predictions based on the training data. 

The line demonstrates the general trend in the data: as years of experience experence increase, salary tends to increase.

## Visualizating the Test Set Results

```python
plt.scatter(X_test, y_test, color = 'red')
plt.plot(X_train, regressor.predict(X_train), color = 'blue')
plt.title('Salary vs. Experience (Test Set)')
plt.xlabel('Years of Experience')
plt.ylabel('Salary')
plt.show()
```

The scatter plot for the test set displays the actual salaies (red dots) against years of experience. The same regression line (blue line) from the training set is plotted. This plot is meant to assess how well the model's predictions generalize unseen data. Ideally, the red dots should be close to the blue line, indicating that the model's predictions align well with the actual test data.
